```python
import os
from PIL import Image
import numpy as np
from skimage.filters import threshold_otsu
```

```python
annotation_folder = 'C:/Users/takatsuto/MLBook/data/edible_and_poisonous_fungi/edible_anno'  # アノテーション画像のフォルダパス
image_folder = 'C:/Users/takatsuto/MLBook/data/edible_and_poisonous_fungi/edible'            # 元画像のフォルダパス
output_folder = 'C:/Users/takatsuto/MLBook/data/edible_and_poisonous_fungi/edible_result3'     # 出力画像の保存先フォルダ


def process_images(annotation_folder, image_folder, output_folder):
    # アノテーション画像と元画像が格納されているフォルダを探索
    annotation_files = os.listdir(annotation_folder)

    for filename in annotation_files:
        # アノテーション画像と元画像のファイルパス
        annotation_path = os.path.join(annotation_folder, filename)
        image_path = os.path.join(image_folder, filename)

        # アノテーション画像と元画像を読み込む
        annotation_img = Image.open(annotation_path).convert('L')  # グレースケール画像
        original_img = Image.open(image_path)

        # アノテーション画像をNumPy配列に変換
        annotation_array = np.array(annotation_img)

        # 大津の二値化でしきい値を計算
        otsu_threshold = threshold_otsu(annotation_array)

        # 元画像をNumPy配列に変換
        original_array = np.array(original_img)

        # 新しい画像の配列を作成（黒背景にする）
        new_array = np.zeros_like(original_array)  # 初期値を黒（0）

        # 大津のしきい値を使って、明るい部分を白として扱い、その部分に元画像を反映
        new_array[annotation_array >= otsu_threshold] = original_array[annotation_array >= otsu_threshold]

        # 結果をPIL画像に戻す
        result_img = Image.fromarray(new_array)

        # 出力先フォルダがなければ作成
        os.makedirs(output_folder, exist_ok=True)

        # 結果画像を保存
        result_img.save(os.path.join(output_folder, filename))

        print(f"Processed {filename}")

# 画像処理を実行
process_images(annotation_folder, image_folder, output_folder)
```
