```python
import os
from PIL import Image

def convert_jpg_to_png(folder_path):
    # フォルダ内のすべてのファイルをリストアップ
    files = os.listdir(folder_path)

    # 対象となる拡張子（.jpg, .jpeg）を指定
    image_extensions = ('.jpg', '.jpeg')

    # 対象となる画像ファイルをフィルタリング
    image_files = [f for f in files if f.lower().endswith(image_extensions)]

    # 画像ファイルを順番に処理
    for filename in image_files:
        old_path = os.path.join(folder_path, filename)

        try:
            # 画像を開く
            with Image.open(old_path) as img:
                # もしCMYKモードなら、RGBモードに変換
                if img.mode == 'CMYK':
                    img = img.convert('RGB')

                # 拡張子を.pngに変更
                new_filename = os.path.splitext(filename)[0] + '.png'
                new_path = os.path.join(folder_path, new_filename)

                # 画像をPNG形式で保存
                img.save(new_path, 'PNG')

                # 元ファイルと新ファイルの情報を表示
                print(f"Converted {filename} to {new_filename}")

            # 画像が閉じた後、古いJPEG/JPGファイルを削除
            os.remove(old_path)
            print(f"Deleted original {filename}")

        except Exception as e:
            print(f"Failed to process {filename}: {e}")

# フォルダパスを指定
folder_path = 'C:/Users/takatsuto/MLBook/data/edible_and_poisonous_fungi/edible'

# JPEG/JPGをPNGに変換
convert_jpg_to_png(folder_path)
```
